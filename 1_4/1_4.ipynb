## Task 1

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import rayleigh
x = np.linspace(rayleigh.ppf(0.01), rayleigh.ppf(0.99), 100)
r = rayleigh.rvs(size=1000)
plt.hist(r, density=True, histtype='stepfilled', alpha=0.2)
rv = rayleigh()
plt.plot(x, rv.pdf(x), 'k-', lw=2, label='frozen pdf')
plt.show()
```

## Task 2

```python
y1 = lambda x: ((x ** 2) ** (1 / 3) + np.sqrt((x ** 4) ** (1 / 3) - 4 * (x ** 2 - 1))) / 2
y2 = lambda x: ((x ** 2) ** (1 / 3) - np.sqrt((x ** 4) ** (1 / 3) - 4 * (x ** 2 - 1))) / 2
x = np.arange(-1.139, 1.14, 0.001)
plt.plot(x, y1(x))
plt.plot(x, y2(x))
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter

path = "dataset_home.txt"
with open(path) as file:
    datalist = [line.replace( '\n', '') for line in file.readlines()]

age = np.array([float(line.split(',')[0]) for line in datalist])
sex = np.array([line.split(',')[1] for line in datalist])
bmi = np.array([float(line.split(',')[2]) for line in datalist])
kids = np.array([int(line.split(',')[3]) for line in datalist])
smoke_flag = np.array([line.split(',')[4] for line in datalist])
area = np.array([line.split(',')[5] for line in datalist])
some_num = np.array([float(line.split(',')[6]) for line in datalist])

plt.scatter(sex, np.log(np.cumsum(age)))
plt.title(" ")
plt.show()

plt.scatter(some_num, age)
plt.title("  ")
plt.show()

counts, bins = np.histogram(age[smoke_flag == 'no'], 46)
no = savgol_filter(counts, 41, 3)
no_smokers = age[smoke_flag == 'no']
plt.hist(no_smokers, bins=45)
plt.title("")
plt.plot(np.arange(np.min(no_smokers),
                   np.max(no_smokers),
                   (np.max(no_smokers)-np.min(no_smokers)) / len(no)), no)
plt.show()
```
